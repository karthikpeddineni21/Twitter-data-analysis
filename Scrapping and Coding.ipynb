# 2.Data Cleaning

```python
# splitting the timestamp
df[['Date','Time']] = df['timestamp'].str.split(' ',expand=True)
df[['Day','Month','Year']] = df['Date'].str.split('/',expand=True)
```

```python
df.head()
```

### 2.1 Removing Duplicate rows and loading required columns

```python
df.drop_duplicates(subset ="text", 
                     keep = False, inplace = True) 
df.to_csv('C:\\Users\\narayana\\Desktop\\msc datascience\\social media analytics\\assesment\\RemDup.csv')
df=pd.read_csv('C:\\Users\\narayana\\Desktop\\msc datascience\\social media analytics\\assesment\\RemDup.csv',usecols={'text','likes','retweets','replies','Year'})
df.head()
```

```python
df['lower']=df['text'].apply(lambda x: " ".join(word.lower() for word in x.split()))
df['lower']=df['lower'].str.replace(r'\W',' ')#removing all the special characters 
df
```

### 2.2 Removing the stop words

```python
from nltk.corpus import stopwords
```

```python
stop_words=stopwords.words('english')# built-in dictionary for removing stop words
```

```python
len(stop_words)
```

```python
# stop words removal
df['stopwords']=df['lower'].apply(lambda x: " ".join(word for word in x.split() if word not in stop_words))
df
```

```python
# Counting all the words in the tweets
Counts=pd.Series(" ".join(df['stopwords']).split()).value_counts()
df10=pd.DataFrame(data=Counts)
df10.head()
```

```python
#Removing the extra words that are not important
other_words=['still','us','go','www','http','https','ly','via','html','u','x','think','tw','nyse','vr','I','e','w','hi','saying','b','oh','the','a','to','and','i','in','it','for','of','on','s','is','they']
df['cleaned_data']=df['lower'].apply(lambda x: " ".join(word for word in x.split() if word not in other_words))
df.head()
```

### 2.3 Lemmatization

```python
from textblob import Word
```

```python
df['lemmatized']=df['cleaned_data'].apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
df.head()
```

# 3.Sentiment analysis

```python
from textblob import TextBlob
```

```python
df['polarity']=df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[0])# for finding polarity
df['subjectivity']=df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[1])# for finding subjectivity
df.drop(['lower','stopwords','cleaned_data'],axis=1,inplace=True)# dropping the unnecessary columns
df.to_csv('C:\\Users\\narayana\\Desktop\\msc datascience\\social media analytics\\assesment\\polaritytweets.csv')
df.head()
```

```python
# Counting likes, tweets and retweets
df['polarity_text']=0
pos_likes=neg_likes=neu_likes=pos_retweets=neg_retweets=neu_retweets=pos_tweets=neg_tweets=neu_tweets=0
y=-1
for x in df[('polarity')]:
    if x>0:
        y=y+1
        df['polarity_text'][y]='positive'
        pos_likes=pos_likes+df['likes'][y]
        pos_retweets=pos_retweets+df['retweets'][y]
        pos_tweets=pos_tweets+1
               
    elif x==0:
        y=y+1
        df['polarity_text'][y]='neutral'
        neu_likes=neu_likes+df['likes'][y]
        neu_retweets=neu_retweets+df['retweets'][y]
        neu_tweets=neu_tweets+1
        
    else:
        y=y+1
        df['polarity_text'][y]='negative'
        neg_likes=neg_likes+df['likes'][y]
        neg_retweets=neg_retweets+df['retweets'][y]
        neg_tweets=neg_tweets+1
        
print('positive tweets =',pos_tweets,' negative tweets =',neg_tweets,' neutral tweets =',neu_tweets)       
print('positive likes =', pos_likes,' negative likes =', neg_likes,' neutral likes =', neu_likes,'positive retweet =', pos_retweets,' negative retweet =', neg_retweets,' neutral retweet =', neu_retweets)
        
        
```
